```python
# single image load example

import matplotlib.pyplot as plt
import cv2

image = cv2.imread(
  '../digits/train/cdf0000_14_3_7.tif', # image path
  cv2.IMREAD_GRAYSCALE                  # read image on grayscale, return an 2D matrix
)

plt.imshow(image)
plt.gray()
plt.show()
image.shape
```

```python
# load all labels into a hash

LABELS_FILE_PATH = '../digits/sample_train.txt' # TODO: change to train base

images_with_labels = {}

with open(LABELS_FILE_PATH) as labels_file:
  for line in labels_file:
    splitted_values = line.split(' ')
    image_name = splitted_values[0]
    label = splitted_values[1].strip()
    images_with_labels[image_name] = label

images_with_labels
```

```python
# flatten flatten values and save a csv with features and labels

import cv2

dataset = []

for key in images_with_labels.keys():
  image_path = f'../digits/train/{key}'
  image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
  normalized_image = image.flatten() / 255.0
  print(normalized_image)
  features = normalized_image.tolist()
  print(len(features))
  label = images_with_labels[key]
  features = features.append(label)
  dataset.append(features)
```

```python
# flatten flatten values and save a csv with features and labels

import cv2

dataset = []

for key in images_with_labels.keys():
  image_path = f'../digits/train/{key}'
  image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
  normalized_image = image.flatten() / 255.0
  print(normalized_image)
  features = normalized_image.tolist()
  print(len(features))
  label = images_with_labels[key]
  features.append(int(label))
  print(features)
  dataset.append(features)
```

```python
import pandas as pd 

dataframe= pd.DataFrame(data=dataset)
dataframe.shape

dataframe.to_csv('train.csv', header=False)
```
